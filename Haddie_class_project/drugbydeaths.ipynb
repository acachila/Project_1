```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
drugdeath = pd.read_csv("Merged_table_of_deathsbydrug_and_region.csv", header = [0,1])
drugdeath.head()
```

```python
c = drugdeath.columns
c
```

```python
drugdeath.columns = drugdeath.columns.droplevel(0)
dd = drugdeath
dd.head()
```

```python
dd.columns = ['Region', 'States', '2013.Nat(oxy, hydrocodone)',
       '2014.Nat(oxy, hydrocodone)',
       '2015.Nat(oxy, hydrocodone)',
       '2016.Nat(oxy, hydrocodone)',
       '2013.Synth(fentanyl,tramadol)',
       '2014.Synth(fentanyl,tramadol)',
       '2015.Synth(fentanyl,tramadol)',
       '2016.Synth(fentanyl,tramadol)', '2013.Methadone',
       '2014.Methadone', '2015.Methadone', '2016.Methadone', '2013.Heroin',
       '2014.Heroin', '2015.Heroin', '2016.Heroin']
```

```python
#First group whole df by drug
deathbyreg = dd.groupby(by = [("Region")]).sum()
deathbyreg.head()
```

```python
deathbyreg_df = pd.DataFrame(deathbyreg)
DBR = deathbyreg_df
DBR
```

```python
meth = DBR.filter(regex="Meth")

synth = DBR.filter(regex="Synth")

heroin = DBR.filter(regex="Heroin")

nat = DBR.filter(regex="Nat")
```

```python
meth
```

```python
# Generate an errorbar plot
# The 'yerr' argument refers to the error bar for each sample, oriented to the y-axis
plt.figure(figsize=(20,10))
plt.errorbar(meth.index, meth["2013.Methadone"], color='k', marker='d', markersize=5, linestyle='dashed', linewidth=1)
plt.errorbar(meth.index, meth["2014.Methadone"], color='g', marker='s', markersize=5, linestyle='dashed', linewidth=1)
plt.errorbar(meth.index, meth["2015.Methadone"], color='b', marker='v', markersize=5, linestyle='dashed', linewidth=1)
plt.errorbar(meth.index, meth["2016.Methadone"], color='r', marker='o', markersize=5, linestyle='dashed', linewidth=1)
x_lim = len(meth.index)
# Chart title
plt.title("Methadone Deaths a 4 year review")
# x label
plt.xlabel("Region")
# y label
plt.ylabel("NUmber of deaths")
# legend 
plt.legend(loc='lower left')
plt.show()
plt.savefig('methadone by year by region.jpeg') 
```
