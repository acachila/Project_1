```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
mw = pd.read_excel("Opiod by Age Group.xlsx", sheet_name = 1)
ne = pd.read_excel("Opiod by Age Group.xlsx", sheet_name = 2)
pw = pd.read_excel("Opiod by Age Group.xlsx", sheet_name = 3)
pl = pd.read_excel("Opiod by Age Group.xlsx", sheet_name = 4)
se = pd.read_excel("Opiod by Age Group.xlsx", sheet_name = 5)
us = pd.read_excel("Opiod by Age Group.xlsx", sheet_name = 6)
```

```python
mw2013 = mw.loc[mw['Year'] == 2013]
mw2014 = mw.loc[mw['Year'] == 2014]
mw2015 = mw.loc[mw['Year'] == 2015]
mw2016 = mw.loc[mw['Year'] == 2016]
```

```python
ne2013 = ne.loc[ne['Year'] == 2013]
ne2014 = ne.loc[ne['Year'] == 2014]
ne2015 = ne.loc[ne['Year'] == 2015]
ne2016 = ne.loc[ne['Year'] == 2016]
```

```python
pw2013 = pw.loc[pw['Year'] == 2013]
pw2014 = pw.loc[pw['Year'] == 2014]
pw2015 = pw.loc[pw['Year'] == 2015]
pw2016 = pw.loc[pw['Year'] == 2016]
```

```python
pl2013 = pl.loc[pl['Year'] == 2013]
pl2014 = pl.loc[pl['Year'] == 2014]
pl2015 = pl.loc[pl['Year'] == 2015]
pl2016 = pl.loc[pl['Year'] == 2016]
```

```python
se2013 = se.loc[se['Year'] == 2013]
se2014 = se.loc[se['Year'] == 2014]
se2015 = se.loc[se['Year'] == 2015]
se2016 = se.loc[se['Year'] == 2016]
```

```python
#midwest
years = [2013, 2014, 2015, 2016]
deaths01 = [mw2013['0-24'].sum(), mw2014['0-24'].sum(), mw2015['0-24'].sum(), mw2016['0-24'].sum()]
deaths02 = [mw2013['25-34'].sum(), mw2014['25-34'].sum(), mw2015['25-34'].sum(), mw2016['25-34'].sum()]
deaths03 = [mw2013['35-44'].sum(), mw2014['35-44'].sum(), mw2015['35-44'].sum(), mw2016['35-44'].sum()]
deaths04 = [mw2013['45-54'].sum(), mw2014['45-54'].sum(), mw2015['45-54'].sum(), mw2016['45-54'].sum()]
deaths05 = [mw2013['55+'].sum(), mw2014['55+'].sum(), mw2015['55+'].sum(), mw2016['55+'].sum()]
plt.figure(num=None, figsize=(10, 7.5), dpi=80)
plt.plot(years, deaths01, color = 'blue', label = '0-24')
plt.plot(years, deaths02, color = 'forestgreen', label = '25-34')
plt.plot(years, deaths03, color = 'red', label = '35-44')
plt.plot(years, deaths04, color = 'black', label = '45-54')
plt.plot(years, deaths05, color = 'gold', label = '55+')
plt.xlim(2013,2016)
plt.xticks(years)
plt.xlabel("Year")
plt.ylabel("Opioid Overdose Deaths")
plt.title("Midwest Region Overdoses by Age Groups")
plt.legend(loc = 'upper left')
plt.savefig('./Output/Midwest.png')

plt.show()
```
