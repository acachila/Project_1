```
%matplotlib inline

import pandas as pd

import matplotlib.pyplot as plt


odtotal = 'Resources//Gender_Total.csv'

gentotal = pd.read_csv(odtotal,thousands=',')

gendf = pd.DataFrame(gentotal)

#Groupby Region
dfreg = gendf.groupby('Region')
dfreg = dfreg.sum()



#Plt the State data for 2013
plt.figure(figsize = (20,10))
State = gendf['State']
Male2013 = gendf['Male2013']/gendf['StateM2013']*100000
Female2013 = gendf['Female2013']/gendf['StateF2013']*100000
plt.bar(State, Male2013)
plt.bar(State, Female2013, color = 'r', width = .3)
plt.xticks(rotation='vertical')
plt.text(33.75,10, '(No Data for 2013)', rotation = '90')
plt.xlim(-.6,50.6)
plt.ylim(0,60)
plt.ylabel('Deaths', fontsize = 25)
plt.yticks(fontsize = 20)
plt.xticks(fontsize = 17)
plt.title('Deaths per 100k in 2013', fontsize = 25)
plt.legend(('Male','Female'), fontsize = 20, loc = 'upper left')
plt.savefig('GenderImages/2013GenderbyState', bbox_inches = 'tight');

#Plt the State data for 2014
plt.figure(figsize = (20,10))
State = gendf['State']
Male2014 = gendf['Male2014']/gendf['StateM2014']*100000
Female2014 = gendf['Female2014']/gendf['StateF2014']*100000
plt.bar(State, Male2014)
plt.bar(State, Female2014, color = 'r', width=.25)
plt.ylabel('Deaths', fontsize = 25)
plt.yticks(fontsize = 20)
plt.xticks(rotation='vertical', fontsize = 17)
plt.xlim(-.6,50.6)
plt.ylim(0,60)
plt.legend(('Male','Female'), fontsize = 20, loc = 'upper left')
plt.title('Deaths per 100k in 2014', fontsize = 25)
plt.savefig('GenderImages/2014GenderbyState', bbox_inches = 'tight');

#Plt the State data for 2015
plt.figure(figsize = (20,10))
State = gendf['State']
Male2015 = gendf['Male2015']/gendf['StateM2015']*100000
Female2015 = gendf['Female2015']/gendf['StateF2015']*100000
plt.bar(State, Male2015)
plt.bar(State, Female2015, color = 'r', width = .25)
plt.xticks(rotation='vertical', fontsize = 17)
plt.ylabel('Deaths',fontsize = 25)
plt.yticks(fontsize = 20)
plt.xlim(-.6,50.6)
plt.ylim(0,60)
plt.legend(('Male','Female'), fontsize = 20, loc = 'upper left')
plt.title('Deaths per 100k in 2015', fontsize = 25)
plt.savefig('GenderImages/2015GenderbyState', bbox_inches = 'tight');

#Plt the State data for 2016
plt.figure(figsize = (20,10))
State = gendf['State']
Male2016 = gendf['Male2016']/gendf['StateM2016']*100000
Female2016 = gendf['Female2016']/gendf['StateF2016']*100000
plt.bar(State, Male2016)
plt.bar(State, Female2016, color = 'r', width = .25)
plt.xticks(rotation='vertical', fontsize = 17)
plt.ylabel('Deaths',fontsize=25)
plt.yticks(fontsize = 20)
plt.xlim(-.6,50.6)
plt.legend(('Male','Female'), fontsize = 20, loc = 'upper left')
plt.title('Deaths per 100k in 2016', fontsize = 25)
plt.savefig('GenderImages/2016GenderbyState', bbox_inches = 'tight');

#Plt the regional data for 2013
plt.figure(figsize = (20,10))
State = gendf['Region']
Male2013 = gendf['Male2013']/gendf['StateM2013']*100000
Female2013 = gendf['Female2013']/gendf['StateF2013']*100000
plt.bar(State, Male2013)
plt.bar(State, Female2013, color = 'r', width = .3)
plt.xticks(rotation='vertical',fontsize = 25)
plt.yticks(fontsize = 20)
plt.xlim(-.6,4.6)
plt.ylim(0,60)
plt.ylabel('Deaths', fontsize = 25)
plt.title('Deaths per 100k in 2013', fontsize = 25)
plt.legend(('Male','Female'), fontsize = 25, loc = 'upper left')
plt.savefig('GenderImages/2013GenderbyRegion', bbox_inches = 'tight');

#Plt the regional data for 2014
plt.figure(figsize = (20,10))
State = gendf['Region']
Male2014 = gendf['Male2014']/gendf['StateM2014']*100000
Female2014 = gendf['Female2014']/gendf['StateF2014']*100000
plt.bar(State, Male2014)
plt.bar(State, Female2014, color = 'r', width = .3)
plt.xticks(rotation='vertical',fontsize = 25)
plt.yticks(fontsize = 20)
plt.xlim(-.6,4.6)
plt.ylim(0,60)
plt.ylabel('Deaths', fontsize = 25)
plt.title('Deaths per 100k in 2014', fontsize = 25)
plt.legend(('Male','Female'), fontsize = 25, loc = 'upper left')
plt.savefig('GenderImages/2014GenderbyRegion', bbox_inches = 'tight');

#Plt the regional data for 2015
plt.figure(figsize = (20,10))
State = gendf['Region']
Male2015 = gendf['Male2015']/gendf['StateM2015']*100000
Female2015 = gendf['Female2015']/gendf['StateF2015']*100000
plt.bar(State, Male2015)
plt.bar(State, Female2015, color = 'r', width = .3)
plt.xticks(rotation='vertical',fontsize = 25)
plt.yticks(fontsize = 20)
plt.xlim(-.6,4.6)
plt.ylim(0,60)
plt.ylabel('Deaths', fontsize = 25)
plt.title('Deaths per 100k in 2015', fontsize = 25)
plt.legend(('Male','Female'), fontsize = 25, loc = 'upper left')
plt.savefig('GenderImages/2015GenderbyRegion', bbox_inches = 'tight');

#Plt the regional data for 2016
plt.figure(figsize = (20,10))
State = gendf['Region']
Male2016 = gendf['Male2016']/gendf['StateM2016']*100000
Female2016 = gendf['Female2016']/gendf['StateF2016']*100000
plt.bar(State, Male2016)
plt.bar(State, Female2016, color = 'r', width = .3)
plt.xticks(rotation='vertical',fontsize = 25)
plt.yticks(fontsize = 20)
plt.xlim(-.6,4.6)
plt.ylim(0,60)
plt.ylabel('Deaths', fontsize = 25)
plt.title('Deaths per 100k in 2016', fontsize = 25)
plt.legend(('Male','Female'), fontsize = 25, loc = 'upper left')
plt.savefig('GenderImages/2016GenderbyRegion', bbox_inches = 'tight');
```
