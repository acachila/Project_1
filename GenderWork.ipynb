```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt


odtotal = 'Resources//Gender_Total.csv'

total = pd.read_csv(odtotal,thousands=',')

df = pd.DataFrame(total)

df.head()
```

```python
df['PercentM']= ((df['Male2016']-df['Male2013'])/df['Male2013'])*100
df['PercentF']= ((df['Female2016']-df['Female2013'])/df['Female2013'])*100
```

```python
dfreg = df.groupby('Region')

dfreg = dfreg.sum()

dfreg.head()
```

```python
maledf = dfreg[['Male2013','Male2014','Male2015','Male2016']].copy()
maledf
```

```python
mwmale = maledf.loc['Midwest']
mwmale

femaledf = dfreg[['Female2013','Female2014','Female2015','Female2016']].copy()
mwfemale = femaledf.loc['Midwest']
mwfemale
```

```python
plt.hlines(0, 0, 10, alpha=0.25)
plt.xlim(0.9,4.1)
plt.ylim(1000,8000)
time = (1,2,3,4)
plt.xticks((time),('2013','2014','2015','2016'))
mwmaleline = plt.plot(time, mwmale,marker ='o', color = 'r')
mwfemaleline = plt.plot(time,mwfemale,marker ='o')
plt.legend(('male', 'female'))
plt.title('Midwest');
```

```python
totalmdf = dfreg[['Male2013','Male2014','Male2015','Male2016','StateM2013','StateM2014','StateM2015','StateM2016']].copy()

totalmdf['perm2013']= (dfreg['Male2013']/dfreg['StateM2013'])*1000000

totalmdf['perm2014']= (dfreg['Male2014']/dfreg['StateM2014'])*1000000

totalmdf['perm2015']= (dfreg['Male2015']/dfreg['StateM2015'])*1000000

totalmdf['perm2016']= (dfreg['Male2016']/dfreg['StateM2016'])*1000000

totalmdf.head()
```

```python
import numpy as np

N = 5
menMeans = (20, 35, 30, 35, 27)
menStd =   (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)
rects1 = ax.bar(ind, menMeans, width, color='royalblue', yerr=menStd)

womenMeans = (25, 32, 34, 20, 25)
womenStd =   (3, 5, 2, 3, 3)
rects2 = ax.bar(ind+width, womenMeans, width, color='seagreen', yerr=womenStd)

# add some
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels( ('G1', 'G2', 'G3', 'G4', 'G5') )

ax.legend( (rects1[0], rects2[0]), ('Men', 'Women') )

plt.show()
```
